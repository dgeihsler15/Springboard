## Imports and Loaded Files

```python
%matplotlib inline
import matplotlib.pyplot as plt 
import pandas as pd
import numpy as np
import json
from pandas.io.json import json_normalize

#load the world bank projects file as a dataframe
df = pd.read_json('world_bank_projects.json')

#load the world bank projects file as a json string
js = json.load((open('world_bank_projects.json')))

```

```python
#Step 1: inspect the data. Please remove the #s on the other print calls to view additional information regarding the dataset.
#print df.info()
#print df.head()
print df.mjtheme_namecode[0]
```

## 1) Top 10 Countries With The Most Projects
While inspecting the World Bank Projects file, you'll find that even though each row of the dataframe contains one country name, there are multiple major project themes for each row in the mjtheme_namecode column. Assuming that the each major project theme represents an indiviual project, we'll have to break out each individual project theme to get an accurate count of projects by country. This is executed in the code below by creating a normalized dataframe (df_total_projects) using mjtheme_namecode and countryname from the JSON string (js) loaded above.

The .value_counts() method is then applied to the countryname column to provide a count for the country names associated with each project. This lists the counts in descending order, so we can access the top 10 countries by slicing the first ten items.

One thing to note is that Africa is one of the country names given in this file. There is not any indication of which countries in Africa fall under this 'countryname'. 

```python
#create normalized dataframe from JSON string
df_total_projects = json_normalize(js, 'mjtheme_namecode', 'countryname')

#count number of projects per country
country_count = df_total_projects.countryname.value_counts()
top_ten_countries = country_count[:10]
print top_ten_countries
```

```python
#plot top ten countries
top_ten_countries.plot(kind='barh').invert_yaxis()
plt.xlabel('Project Count')
plt.title('Top 10 Countries With Most Projects')
plt.show()
```

## 2) Top 10 Major Project Themes: 
Like the first excercise, we will need to break out each individual project theme to find the top 10 major project themes. We will utilize the df_total_projects dataframe created previously. Inspecting the project themes more closely, you'll notice that all of the project codes are filled in, but some of the names are missing. With this in mind, we can run the .value_counts() method on the code column to count the numer of major product themes across all of the projects.

We'll then take out the first 10 items from theme_count, and find our top ten major project theme codes. 

```python
#count number of projet theme codes
theme_count = df_total_projects.code.value_counts()
top_ten_themes = theme_count[:10]
print top_ten_themes
```

If we want to view the top ten project themes with the associated theme names, we can first fill in the missing values for the name column (this work is done in excercise 3 below). Then we can use the .value_counts() method on the name column of the updated dataframe (df_updated) to count the top major theme names. 




## 3) DataFrame with missing project theme names filled in: 
As mentioned in excercise 2, not all of the project theme names are filled in. We can start by sorting the df_total_projects dataframe by 'code' so that the missing name values can be filled in appropriately. This can be done using the .sort_values() method on the code column of the dataframe. Noticing that the 'code' column contains string values, we can use the astype method on the 'code' column to change the values to integers. The dataframe can then be sorted by 'code' correctly.

One way to fill in missing values in a pandas series is to use the .fillna() method. As the current missing values are empty strings, we can first replace those values with NaNs, and then use the .fillna() method on the 'name' column of the sorted dataframe (df_updated). 

We can now use the .value_counts() method on the updated dataframe to count the number of project themes by name.

Finally we can check to make sure all the fields in the name column were filled in with the .info() method on the new dataframe (df_updated).



```python
#change values to integers and sort. In ascending order, the first value is blank. Reverse order to make ffill more effective.  
df_total_projects['code'] = df_total_projects['code'].astype(int)
df_updated = df_total_projects.sort_values('code', ascending=False)

#replace the blank values with NaNs
sorted_nan = df_updated.replace('', np.nan, regex=True)

#fill in missing values using forward fill
df_updated['name'] = sorted_nan.name.fillna(method='ffill')

#count the top ten themes by name and plot results
top_ten_names = df_updated.name.value_counts()[:10]
top_ten_names.plot(kind='barh').invert_yaxis()
plt.xlabel('Count of Major Project Themes')
plt.title('Top 10 Project Themes')
```
